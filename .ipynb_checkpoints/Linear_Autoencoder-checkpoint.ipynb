```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('./anonymized_data.csv')
ds.head()
```

```python
data_x = ds.iloc[:, :-1]
data_y = ds.iloc[:, -1]
```

```python
data_x.head()
```

```python
data_y.head()
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_x = scaler.fit_transform(data_x)
```

```python
scaled_x[0]
```

```python
scaled_x.shape
```

## Autoencoder

```python
import tensorflow as tf
```

```python
num_inputs = scaled_x.shape[1]
hidden_inputs = 2
num_outputs = num_inputs
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(num_inputs, activation=None))
model.add(Dense(hidden_inputs, activation=None))
model.add(Dense(num_outputs, activation=None))
```

```python
model.compile(optimizer='adam', loss='MSE',metrics=['accuracy'])
```

```python
model.fit(scaled_x,scaled_x, batch_size=32, epochs=10, validation_data=[scaled_x,scaled_x])
```

```python
model.summary()
```

## Visualize

```python
import tensorflow.keras.backend as K
```

```python
outputs = K.function([model.layers[0].input], [model.layers[1].output])
```

```python
layer_output = outputs([scaled_x])[0]
```

```python
layer_output.shape
```

```python
layer_output
```

```python
plt.scatter(layer_output[:,0], layer_output[:,1],c=data_y)
```

#### Another way to visualize

```python
from tensorflow.keras.models import Model
```

```python
hidden_layer = Model(inputs=model.input, outputs=model.get_layer(index=1).output)
```

```python
hidden_output = hidden_layer.predict(scaled_x)
```

```python
hidden_output.shape
```

```python
plt.scatter(hidden_output[:,0], hidden_output[:,1],c=data_y)
```
